```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinic_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
trial_data = mouse_data.merge(clinic_data, how='outer')

# Display the data table for preview
trial_data.head()


```

## Plot Helpers

```python
# Get the name of each drug & only select that data
drugs = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
trial_data= trial_data.loc[trial_data["Drug"].isin(drugs)]

# get the colors (blue, green, red, black)
colors = ['r','b','g','k']

# get the markers (triangle, square, circle, thin diamond)
markers = ['o','^','s','d']
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume = trial_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
mean_tumor_volume = mean_tumor_volume.unstack(level = 0)

# Preview DataFrame
mean_tumor_volume.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_volume = trial_data.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
sem_tumor_volume = sem_tumor_volume.unstack(level=0)

# Preview DataFrame
sem_tumor_volume.head()
```

```python
# Generate the Plot (with Error Bars)
for i in range (0, len(drugs)):
    plt.errorbar(x=mean_tumor_volume.index, y=mean_tumor_volume.iloc[:,i],
                 yerr=sem_tumor_volume.iloc[:,i],marker=markers[i], color=colors[i],
                linestyle=':', linewidth=1, label=drugs[i])

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend()
plt.grid()

# Save
plt.savefig("./plots/Tumor Response to Treatment.png")

# Show
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_site = trial_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
mean_met_site = mean_met_site.unstack(level = 0)

# Preview DataFrame
mean_met_site.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_met_site = trial_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
sem_met_site = sem_met_site.unstack(level = 0)

# Preview DataFrame
sem_met_site.head()
```

```python
for i in range (0, len(drugs)):
    plt.errorbar(x=mean_met_site.index, y=mean_met_site.iloc[:,i],
                 yerr=sem_met_site.iloc[:,i],marker=markers[i], color=colors[i],
                linestyle=':', linewidth=1, label=drugs[i])

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met Sites")
plt.legend()
plt.grid()

# Save
plt.savefig("./plots/Metastatic Spread During Treatment.png")

# Show
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_numbers = trial_data.groupby(["Drug", "Timepoint"])["Mouse ID"].count()

# Convert to DataFrame
mice_numbers = mice_numbers.unstack(level = 0)

mice_numbers = mice_numbers.divide(mice_numbers.ix[0]/100)

# Preview DataFrame
mice_numbers.head()
```

```python
# Generate the Plot (Accounting for percentages)
for i in range (0, len(drugs)):
    plt.errorbar(x=mice_numbers.index, y=mice_numbers.iloc[:,i],marker=markers[i], color=colors[i],
                linestyle=':', linewidth=1, label=drugs[i])

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend()
plt.grid()             
             
# Save the Figure
plt.savefig("./plots/Survival During Treatment.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
data = trial_data.loc[trial_data["Timepoint"]==45].drop(columns=["Metastatic Sites", "Timepoint"])
data["pct_change"]=(data["Tumor Volume (mm3)"]-45)*100/45
pct_change = data.groupby(["Drug"]).mean()["pct_change"]
```

```python
def autolabel(rects, ax):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for rect in rects:
        height = rect.get_height()

        # Fraction of axis height taken up by this rectangle
        p_height = (height / y_height)

        # If we can fit the label above the column, do that;
        # otherwise, put it inside the column.
        if p_height > 0.50: # arbitrary; 95% looked good to me.
            label_position = height - (y_height * 0.05)
        else:
            label_position = height + (y_height * 0.01)

        ax.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
# Store all Relevant Percent Changes into a Tuple
values = tuple(pct_change)

# Orient widths. Add labels, tick marks, etc. 
x_axis = np.arange(len(values))

# Use functions to label the percentages of changes
tick_locations = [value for value in x_axis]

# Call functions to implement the function calls
fig, ax = plt.subplots()
f = plt.bar(x_axis, values, color=['g','r','r','r'])
autolabel(f, ax)
plt.xticks(tick_locations, drugs)
plt.grid()
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

# Save the Figure
plt.savefig("./plots/Tumor Change Over 45 Day Treatment")

# Show the Figure
plt.show()
```
